```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geoplot
import geopandas as gpd
import contextily as ctx
import json
%matplotlib inline
from shapely.geometry import Point, shape
from folium.plugins import HeatMap
```

```python
parks_df=pd.read_csv('Department_of_Recreation_and_Parks__Facility_and_Park_Information.csv')
epl_df=pd.read_csv('Emergency_Preparedness_Locations_-_2018.csv')
```

```python
parks_df=parks_df.iloc[np.where(parks_df.LocationType.isin(['Beaches', 'Dog Parks', 'Equestrian Centers', 'Golf Courses', 'Gardens', 'Open Space', 'Parks']))]
parks_df.iloc[np.where(parks_df.LocationType.isin(['Open Space']))]
```

```python
A=epl_df[['Category', 'Name', 'Latitude', 'Longitude']]
B=parks_df[['LocationType', 'Location Name', 'GeoLat', 'GeoLong']]
B=B.rename(columns=dict(zip(B.columns, A.columns)))
df=A.append(B, ignore_index=True)
df.sample(3)
```

```python
gdf=gpd.GeoDataFrame(df, geometry=[Point(xy) for xy in zip(df.Longitude, df.Latitude)], crs={'init': 'epsg:4326'})
```
